```python
import sys
sys.path.append('..')
from nx_graph.utils_test import create_trained_model
from nx_graph.utils_test import plot_metrics
```

```python
config_file = '../configs/nxgraph_test_NOINT.yaml'
input_ckpt = '../trained_results/nxgraph_big_test_NOINT/bak'
model = create_trained_model(config_file, input_ckpt)
```

```python
%%time
iteration = 7128
odd, tdd = model(iteration, 32)
plot_metrics(odd, tdd, odd_th=0.5)
```

```python
import sklearn.metrics
import matplotlib.pyplot as plt
```

```python
p, r, t = sklearn.metrics.precision_recall_curve(tdd, odd)
plt.figure()
plt.plot(t, p[:-1], label='purity')
plt.plot(t, r[:-1], label='efficiency')
plt.xlabel('Cut on model score')
plt.legend(loc=0)
```

```python
plt.figure()
plt.plot(p, r, )
plt.xlabel('Purity')
plt.ylabel('Efficiency')
```

```python
import pickle
```

```python
pickle.dump([p, r], open("purity_recall.pkl", 'wb'))
```
